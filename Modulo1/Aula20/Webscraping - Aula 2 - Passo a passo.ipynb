```python
# Generated by Selenium IDE
# import pytest -> ELIMINAR OU COMENTAR O IMPORT PYTEST - NÃO VAMOS UTILIZAR
import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
```

```python
from bs4 import BeautifulSoup
```

```python
class TestSearchclick():
    def __init__(self): # troquei para o método construtor
        self.driver = webdriver.Chrome('chromedriver.exe') # ou 'C:/Users/Renata/DSdegree/Aula20/chromedriver.exe'
        self.vars = {}

    def teardown_method(self):
        self.driver.quit()

    def test_searchclick(self):
        self.driver.get("https://www.google.com/")
        self.driver.set_window_size(1552, 840)
        self.driver.find_element(By.NAME, "q").click()
        self.driver.find_element(By.NAME, "q").send_keys("python")
        self.driver.find_element(By.NAME, "q").send_keys(Keys.ENTER)
#         self.driver.find_element(By.NAME, "q").send_keys("python")
#         self.driver.find_element(By.NAME, "q").send_keys(Keys.ENTER)
```

```python
teste = TestSearchclick()

teste.test_searchclick()
```

```python
url = teste.driver.page_source
site = BeautifulSoup(url, 'html.parser')
```

```python
print(site.prettify())
```

```python
site.find_all(class_="LC20lb DKV0Md") # DEVOLVE TODAS AS OCORRÊNCIAS DENTRO DE UMA LISTA
```

```python
# Generated by Selenium IDE
import pytest
import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities

class TestCetipdados():
    def __init__(self): # troquei para o método construtor
        self.driver = webdriver.Chrome('chromedriver.exe') # ou 'C:/Users/Renata/DSdegree/Aula20/chromedriver.exe'
        self.vars = {}

    def teardown_method(self):
        self.driver.quit()
  
    def test_cetipdados(self):
        self.driver.get("http://estatisticas.cetip.com.br/astec/series_v05/paginas/lum_web_v05_template_informacoes_di.asp?str_Modulo=completo&int_Idioma=1&int_Titulo=6&int_NivelBD=2")
        self.driver.set_window_size(1552, 840)
        self.driver.find_element(By.NAME, "DT_DIA_DE").click()
        self.driver.find_element(By.NAME, "DT_DIA_DE").clear()
        self.driver.find_element(By.NAME, "DT_DIA_DE").send_keys("01")
        self.driver.find_element(By.NAME, "DT_MES_DE").click()
        self.driver.find_element(By.NAME, "DT_MES_DE").clear()
        self.driver.find_element(By.NAME, "DT_MES_DE").send_keys("05")
        self.driver.find_element(By.NAME, "DT_ANO_DE").click()
        self.driver.find_element(By.NAME, "DT_ANO_DE").clear()
        self.driver.find_element(By.NAME, "DT_ANO_DE").send_keys("2020")
        self.driver.find_element(By.NAME, "DT_DIA_ATE").click()
        self.driver.find_element(By.NAME, "DT_DIA_ATE").clear()
        self.driver.find_element(By.NAME, "DT_DIA_ATE").send_keys("30")
        self.driver.find_element(By.NAME, "DT_MES_ATE").click()
        self.driver.find_element(By.NAME, "DT_MES_ATE").clear()
        self.driver.find_element(By.NAME, "DT_MES_ATE").send_keys("06")
        self.driver.find_element(By.NAME, "DT_ANO_ATE").click()
        self.driver.find_element(By.NAME, "DT_ANO_ATE").clear()
        self.driver.find_element(By.NAME, "DT_ANO_ATE").send_keys("2020")
        self.driver.find_element(By.CSS_SELECTOR, ".button:nth-child(2)").click()  
```

```python
teste_cetip = TestCetipdados()

teste_cetip.test_cetipdados()
```
